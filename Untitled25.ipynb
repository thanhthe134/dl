```python
import numpy as np
from collections import defaultdict
import math

def train_gaussian_nb(X_train, y_train):
    model = {}
    labels = np.unique(y_train)

    for label in labels:
        idxs = np.where(y_train == label)[0]
        X_label = X_train[idxs]
        model[label] = {
            'mean': X_label.mean(axis=0),
            'var': X_label.var(axis=0) + 1e-6,  # tránh chia cho 0
            'prior': len(idxs) / len(X_train)
        }

    return model
```

```python
def gaussian_prob(x, mean, var):
    exponent = np.exp(-((x - mean) ** 2) / (2 * var))
    return (1.0 / np.sqrt(2 * np.pi * var)) * exponent
```

```python
def predict_gaussian_nb(model, X_test):
    preds = []

    for x in X_test:
        probs = {}
        for label, stats in model.items():
            log_prob = math.log(stats['prior'])
            mean = stats['mean']
            var = stats['var']
            for i in range(len(x)):
                prob = gaussian_prob(x[i], mean[i], var[i])
                log_prob += math.log(prob + 1e-9)  # thêm epsilon để tránh log(0)
            probs[label] = log_prob
        preds.append(max(probs, key=probs.get))

    return np.array(preds)
```

```python
# Chuyển về numpy
X_train_np = X_train.to_numpy()
X_test_np = X_test.to_numpy()
y_train_np = y_train.to_numpy()
y_test_np = y_test.to_numpy()

# Huấn luyện và dự đoán
model = train_gaussian_nb(X_train_np, y_train_np)
y_pred = predict_gaussian_nb(model, X_test_np)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("Accuracy:", accuracy_score(y_test_np, y_pred))
print("Confusion matrix:\n", confusion_matrix(y_test_np, y_pred))
print("Classification report:\n", classification_report(y_test_np, y_pred))
```

```python
def predict_gaussian_nb_with_threshold(model, X_test, threshold=0.5):
    preds = []

    for x in X_test:
        probs = {}
        for label, stats in model.items():
            log_prob = math.log(stats['prior'])
            mean = stats['mean']
            var = stats['var']
            for i in range(len(x)):
                prob = gaussian_prob(x[i], mean[i], var[i])
                log_prob += math.log(prob + 1e-9)
            probs[label] = log_prob

        # Chuyển log-probs về xác suất thực
        max_log = max(probs.values())
        exp_probs = {k: math.exp(v - max_log) for k, v in probs.items()}
        total = sum(exp_probs.values())
        normalized_probs = {k: v / total for k, v in exp_probs.items()}

        # Áp threshold: ví dụ ưu tiên class = 1 (tái phạm)
        if normalized_probs.get(1, 0) >= threshold:
            preds.append(1)
        else:
            preds.append(0)

    return np.array(preds)
```

```python
threshold = 0.04

y_pred_thres = predict_gaussian_nb_with_threshold(model, X_test_np, threshold)

print(f"Accuracy với threshold={threshold}: {accuracy_score(y_test_np, y_pred_thres):.4f}")
print("Confusion matrix:\n", confusion_matrix(y_test_np, y_pred_thres))
print("Classification report:\n", classification_report(y_test_np, y_pred_thres))
```
